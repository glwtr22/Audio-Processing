```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs, data = wavfile.read("/Users/happy/resource/Tone500Hz_short.wav")
```

```python
print(fs, data.shape)
print(data)
times = np.arange(len(data))/float(fs)
```

```python
plt.fill_between(times, data)
plt.xlim(times[0], times[-1])
plt.xlabel("times(s)")
plt.ylabel("amplitude")
plt.figure(figsize = (20,5))
plt.show()
```

```python
len(data)
```

```python
l_arr = data
r_arr = data
```

```python
stereo_arr = []
for l, r in zip(l_arr, r_arr):
    stereo_arr.append(l)
    stereo_arr.append(r)
```

```python
len(stereo_arr)
```

```python
print(stereo_arr)
```

```python
stereo_arr = np.array(stereo_arr, dtype=np.float32)
wavfile.write("stereo_transformed.wav", 3200, stereo_arr)
```

```python
times2 = np.arange(len(stereo_arr))/float(fs)

plt.fill_between(times2, data)
plt.xlim(times2[0], times2[-1])
plt.xlabel("times(s)")
plt.ylabel("amplitude")
plt.figure(figsize = (20,5))
plt.show()
```

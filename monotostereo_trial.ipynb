```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt

FIG_SIZE = (20, 5)
```

```python
file = "/Users/happy/resource/Tone500Hz_short.wav"
```

```python
sig, sr = librosa.load(file)
```

```python
print(sig, sig.shape)
```

```python
plt.figure(figsize = FIG_SIZE)
librosa.display.waveshow(sig, sr, alpha = 0.5)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

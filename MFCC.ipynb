```python
import librosa, librosa.display
import matplotlib.pyplot as plt
```

###### 아 음성

```python
file = "/Users/happy/resource/a.wav"

signal, sample_rate = librosa.load(file, sr = 8000)
mfcc12 = librosa.feature.mfcc(y = signal, sr = sample_rate, n_mfcc = 12)
mfcc24 = librosa.feature.mfcc(y = signal, sr = sample_rate, n_mfcc = 24)
print(sample_rate)
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(mfcc12, x_axis ='time', ax = ax)
fig.colorbar(img, ax = ax)
ax.set(title = 'MFCC-12')
```

```python
fig, ax = plt.subplots()
img = librosa.display.specshow(mfcc24, x_axis ='time', ax = ax)
fig.colorbar(img, ax = ax)
ax.set(title = 'MFCC-24')
```

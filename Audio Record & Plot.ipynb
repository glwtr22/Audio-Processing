```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
```

```python
FIG_SIZE = (15, 10)
file_a = "/Users/happy/resource/a.wav"

signal_a, sample_rate_a = librosa.load(file_a, sr = 8000)
print('signal shape : ', signal_a.shape)
```

```python
plt.figure(figsize = FIG_SIZE)
librosa.display.waveshow(signal_a, sample_rate_a, alpha = 0.4)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform : Ah")
```

```python
start = len(signal_a)/4
end = len(signal_a)*5/8
signal_a_cut = signal_a[round(start):round(end)]
time2 = np.linspace(0, len(signal_a_cut)/sample_rate_a, len(signal_a_cut))
plt.figure(figsize = FIG_SIZE)
librosa.display.waveshow(signal_a_cut, sample_rate_a, alpha = 0.4)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform : Trimed A")
```

```python
#  
a_fft = np.fft.fft(signal_a_cut)
print('fft shape : ', a_fft.shape)
```

```python
a_spectrum = np.abs(a_fft)
print("spectrum shape : ", a_spectrum.shape)
```

```python
a_f = np.linspace(0, sample_rate_a, len(a_spectrum))
print("af shape : ", a_f.shape)
```

```python
a_left_spectrum = a_spectrum[:int(len(a_spectrum)/2)]
a_left_f = a_f[:int(len(a_spectrum)/2)]
print("left_spectrum shape : ", a_left_spectrum.shape)
```

```python
plt.figure(figsize = FIG_SIZE)
plt.plot(a_left_f, a_left_spectrum, alpha=0.4)
plt.xlabel("Fequency")
plt.ylabel("Magnitude")
plt.title("Power spectrum : A")
```
